```python
import numpy as np
import pandas as pd
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
```

```python
import pickle
```

```python
df = pd.read_csv("encoded_cleaned_train_df.csv")
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train_h1n1 \
, y_test_h1n1, y_train_seasonal, y_test_seasonal \
      = train_test_split(df.drop(['h1n1_vaccine','seasonal_vaccine','respondent_id'], axis = 1)
                                , df['h1n1_vaccine'], df['seasonal_vaccine']
                                , test_size=0.2, random_state=42)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
```

```python
models,predictions = clf.fit(X_train, X_test, y_train_h1n1, y_test_h1n1)
```

```python
models
```

```python
from sklearn.neighbors import NearestCentroid
```

```python
# Create a NearestCentroid classifier
clf1 = NearestCentroid()

# Define the parameter grid for grid search
param_grid = {
    'shrink_threshold': [None, 0.1, 0.5, 1.0]
}

# Perform grid search
grid_search = GridSearchCV(clf1, param_grid, cv=5)
grid_search.fit(X_train, y_train_h1n1)

# Print the best parameters and score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

```python
# Create a NearestCentroid classifier
clf2 = AdaBoostClassifier()

# Fit the classifier to the training data
clf2.fit(X_train, y_train_h1n1)

# Predict on the test data
y_pred_h1n1 = clf2.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test_h1n1, y_pred_h1n1)

# Print accuracy
print("Accuracy:", accuracy)
print()
print(classification_report(y_test_h1n1, y_pred_h1n1))
```

```python
# Export the clf2 object
with open("h1n1_classifier", 'wb') as file:
    pickle.dump(clf2, file)
```

```python
models1,predictions1 = clf.fit(X_train, X_test, y_train_seasonal, y_test_seasonal)
models1
```

```python
# Create an AdaBoostClassifier
adaboost = AdaBoostClassifier()

# Define the parameter grid for grid search
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.1, 0.5, 1.0]
}

# Perform grid search
grid_search = GridSearchCV(adaboost, param_grid, cv=5)
grid_search.fit(X_train, y_train_seasonal)

# Print the best parameters and score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

```python
# Use the best estimator for prediction
best_estimator = grid_search.best_estimator_
y_pred_seasonal = best_estimator.predict(X_test)

# Print classification report
print("Classification Report:")
print(classification_report(y_test_seasonal, y_pred_seasonal))
```
